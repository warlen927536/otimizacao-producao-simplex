```python
import numpy as np
import matplotlib.pyplot as plt
from pulp import LpMaximize, LpProblem, LpVariable, value

# 1. Configuração do Modelo (Igual ao Excel)
model = LpProblem(name="Otimizacao_Producao", sense=LpMaximize)
x = LpVariable(name="Produto_A", lowBound=0)
y = LpVariable(name="Produto_B", lowBound=0)

model += 150 * x + 100 * y  # Lucro
model += (5 * x + 3 * y <= 500)  # Matéria-prima
model += (2 * x + 4 * y <= 400)  # Mão de Obra
model += (3 * x + 2 * y <= 450)  # Máquina

model.solve()

# 2. Preparando os dados para o gráfico
d = np.linspace(0, 150, 300)
plt.figure(figsize=(10, 7))

# Criando as linhas das restrições
plt.plot(d, (500 - 5*d)/3, label='Matéria-prima (5x + 3y <= 500)', color='red', linestyle='--')
plt.plot(d, (400 - 2*d)/4, label='Mão de Obra (2x + 4y <= 400)', color='blue', linestyle='--')
plt.plot(d, (450 - 3*d)/2, label='Máquina (3x + 2y <= 450)', color='green', linestyle='--')

# Configurações do gráfico
plt.xlim(0, 110)
plt.ylim(0, 110)
plt.xlabel('Quantidade Produto A')
plt.ylabel('Quantidade Produto B')
plt.title('Análise de Otimização Linear - Região de Viabilidade', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)

# Pintando a área permitida (Região de Viabilidade)
y_limit = np.minimum(np.minimum((500 - 5*d)/3, (400 - 2*d)/4), (450 - 3*d)/2)
plt.fill_between(d, np.maximum(0, y_limit), where=(d>=0), color='gray', alpha=0.2, label='Área Viável')

# Marcando o ponto máximo (A solução do Solver)
plt.plot(value(x), value(y), 'ko', markersize=10)
plt.annotate(f'PONTO ÓTIMO\n(A={value(x):.1f}, B={value(y):.1f})', 
             xy=(value(x), value(y)), xytext=(value(x)+10, value(y)+10),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()

print(f"Lucro Total Máximo: R$ {value(model.objective):.2f}")
```
